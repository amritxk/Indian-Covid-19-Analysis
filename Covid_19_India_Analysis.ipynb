```python
#Dropping unnecessary columns from Covid_Df
covid_df.drop(['Sno' , 'Time','ConfirmedIndianNational','ConfirmedForeignNational'], axis=1 , inplace=True)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.info()
```

```python
#Creating a new ActiveCases Column 
covid_df['ActiveCases'] = covid_df['Confirmed'] -(covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
#Creating a Pivot
statewise = pd.pivot_table(covid_df , values =['Confirmed' ,'Deaths' ,'Cured'], index='State/UnionTerritory',aggfunc= max )
```

```python
#Creating Recoveryrate and Mortality Rate Columns
statewise['RecoveryRate'] = statewise['Cured']*100 / statewise['Confirmed']
statewise['MortalityRate'] = statewise['Deaths']*100 / statewise['Confirmed']
```

```python
#Sorting the Values according to Confirmed Cases in Desc
statewise = statewise.sort_values(by='Confirmed' , ascending=False)
```

```python
#Plotting the Pivot Table
statewise.style.background_gradient(cmap='Spectral')
```

```python
#Top 10 active cases states
top_10_active_cases=covid_df.groupby(by='State/UnionTerritory').max()[['ActiveCases' , 'Date']].sort_values(by=['ActiveCases'], ascending = False).reset_index()
top_10_active_cases=top_10_active_cases.iloc[:10]
```

```python
fig = plt.figure(figsize=(13,9))
plt.title('Top 10 States with Most Active Cases')
ax=sns.barplot( data= top_10_active_cases , y='ActiveCases' , x='State/UnionTerritory', palette='rocket')
plt.xlabel('States')
plt.ylabel('TotaL Active Cases')
plt.show()

```

```python
#Top 10 States with Highest Deaths
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by='Deaths',ascending = False).reset_index().iloc[:12]
top_10_deaths.drop([1,3],axis=0,inplace=True)
```

```python
fig = plt.figure(figsize=(14,9))
plt.title('Top 10 States with Most Deaths')
ax=sns.barplot( data= top_10_deaths , y='Deaths' , x='State/UnionTerritory', palette='mako')
plt.xlabel('States')
plt.ylabel('TotaL Deaths')
plt.show()
```

```python
# Active Cases Growth rate
fig = plt.figure(figsize=(14,9))

ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(top_10_active_cases['State/UnionTerritory'].iloc[:5])],x='Date',y='ActiveCases',hue='State/UnionTerritory')
plt.title('Top 5 Affected States')
plt.xlabel('Date')
plt.ylabel('Total Active Cases')
```

```python
# Removing rows where state is mentioned as INDIA
vaccine = vaccine_df[vaccine_df['State']!='India']
vaccine
```

```python
#Most Vaccinated State
max_vaccine = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total')
```

```python
max_vaccine = max_vaccine.sort_values('Total', ascending=False)[:5]
```

```python
fig = plt.figure(figsize=(14,9))
plt.title('Top 5 States with Most Vaccinations')
ax=sns.barplot( data= max_vaccine , y='Total' , x=max_vaccine.index, palette='mako')
plt.xlabel('States')
plt.ylabel('Toatl Individual Vaccinated')
plt.show()
```

```python
#Least 5 Vaccinated States
min_vaccine = vaccine.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total')
min_vaccine = min_vaccine.sort_values('Total', ascending=True)[:5]
```

```python
fig = plt.figure(figsize=(14,9))
plt.title('Least 5 States with Vaccinations')
ax=sns.barplot( data= min_vaccine , y='Total' , x=min_vaccine.index, palette='mako')
plt.xlabel('States')
plt.ylabel('Toatl Individual Vaccinated')
plt.show()
```
